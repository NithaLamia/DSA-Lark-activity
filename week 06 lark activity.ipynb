3.On average which region has the highest


a. GDP per capita (current US$)

```python
data1=data[['Region','GDP per capita (current US$)']].groupby('Region').mean()
data1=data1.sort_values('GDP per capita (current US$)',ascending=True)
data1
```

```python
data1.nlargest(1,['GDP per capita (current US$)'])
```

b. International trade: Imports (million US$)

```python
data2=data[['Region','International trade: Imports (million US$)']].groupby('Region').mean()
data2=data2.sort_values('International trade: Imports (million US$)',ascending=False)
data2
```

```python
data2.nlargest(1,['International trade: Imports (million US$)'])
```

4. Find out which region has more consistent with respect to GDP per capita
(current US$)

```python
data1=data[['Region','GDP per capita (current US$)']].groupby('Region').std()
data1=data1.sort_values('GDP per capita (current US$)',ascending=True)
data1
```
